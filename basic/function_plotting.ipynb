# Plotting a function

In this example we show how to plot a function using [matplotlib](https://matplotlib.org/), a free python plotting module that is built into Jupyter. Matplotlib can be used to make a huge variety of plots, some example of which (along with source code) can be found in the [matplotlib gallery](https://matplotlib.org/gallery/index.html). Here we'll demonstrate the basics so that you can make simple plots of functions you want to investigate.

To use the plotting functions we have to "import" them from their module, which is [matplotlib.pyplot](https://matplotlib.org/api/pyplot_summary.html). Importing a function makes it available for you to use in your own programs. We import the whole module, including the function [`plot`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot), like so:

```python
import matplotlib.pyplot as plt
```

We'll also import the [`numpy`](http://www.numpy.org/) module, which is full of useful functions that can help us quickly achieve results. It's used so much in scientific programming that you will often see people shorten its name to `np` when importing just to save a few keystrokes. To do this use the `as` keyword when importing:

```python
import numpy as np
```

## Creating the plot

At a bare minimum, the `plot` function requires two inputs, a list of $x$ values and $y$ values. In order to make the plot we must create these. First the x-values are chosen to be a vector linearly spaced between -10 and 10 using the [`linspace`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html) function from numpy.

```python
xvals = np.linspace(-10,10)
```

Now create the $y$-values by applying a function to the $x$-values. Numpy has most of the most common maths functions for us to use.

```python
yvals = np.sin(xvals)
```

Finally we are ready to plot. We use the function [`plot()`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) from matplotlib.

```python
plt.plot(xvals, yvals)
```

### Decorating the plot
Of course, a plot with no axis labels is next to useless, so let's do a bit better. We can add axis labels, a grid and a title using the decoration functions `xlabel`, `ylabel`, `grid` and `title`.

```python
plt.plot(xvals, yvals)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('A Sine Wave')
```

Plotting many curves
===

Very nice, but what if we want to plot a few different curves to compare them. Things can get confusing quickly.
For example, let's say we want to study the potential energy stored in a spring as it is extended

$E=\frac{1}{2}kx^2$

We might want to compare different springs with different spring constants $k_1$ and $k_2$. To make life a bit easier we can write a function (called `E_spring()`) to compute the energy for an arbitrary spring constant. Notice that in the documentation we specify the units expected. This is good practice and prevents confusion when you want to reuse a function later.  The  `"""` lines mark the start and the end of comments within the code.

```python
def E_spring(k,x):
    """
    Energy in an extended spring with spring constant k.
    Arguments:
        k : spring constant [N/m]
        x : extension [m]
    Returns:
        Energy = (1/2) * k * x^2 [J]
    """
    return 0.5*k*x**2
```

Now let's compute the potential energy for a couple of different spring constants, 1000 N/m and 2000 N/m. Let the spring extend up to 0.1m.

```python
x = np.linspace(0,0.1)
E1000 = E_spring(1000, x)
E2000 = E_spring(2000, x)
```

We can plot multiple energy functions by calling the `plot()` function once for each one. To make it easy to tell the functions apart we will add a legend too. To make a legend we must give the `label` keyword as an additional argument to `plot()`, then call the [`legend()`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend) function to display it.

```python
plt.plot(x, E1000, label='k=1000 N/m')
plt.plot(x, E2000, label='k=2000 N/m')
plt.xlabel('Extension [meters]')
plt.ylabel('Potential Energy [Joules]')
plt.title('Spring energy')
plt.grid(True)
plt.legend()
```

## Polish

We can add a few finishing touches. Here's an example that tightens up the axis limits so there is less of a gap, and shows how to choose a particular colour or line style for the lines. They keywords are fairly self-explanatory. Try experimenting with different styles for yourself. For a full list of the different plotting options take a look at the [documentation for `plot()`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot).

Finally, we can also save the plot to disk using [`savefig()`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.savefig.html#matplotlib.pyplot.savefig). Savefig is clever enough to allow several different file extensions, the most commonly used in reports being `.png` and `.pdf`.  The [file](../files/springs.png) will appear in the jupyter folder alongside the .ipynb file that created it. You can download it from there.

```python
# Import a bunch of plotting functions
from matplotlib.pyplot import plot, xlabel, ylabel, title, grid, xlim, ylim, legend, savefig

# Plot with a red line of width 3
plot(x, E1000, label='k=1000 N/m', color='red', linewidth=3)
# Plot with a dash-dotted line with width 5
plot(x, E2000, label='k=2000 N/m', linestyle='-.', linewidth=5)

xlabel('Extension [meters]')
ylabel('Potential Energy [Joules]')
title('Spring energy')
grid()

# Set the axis limits
xlim( [0, 0.1] )
ylim( [0, 10] )

# Make the legend font a little bit bigger
legend(fontsize=12)

# Save the figure
savefig('springs.png')
```
