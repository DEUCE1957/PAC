## An example
We will set ourselves the task of plotting an exponentially-decaying sine wave, ie 

$$f(x) = \exp(-t/\tau)\sin 2\pi f t,$$ 

as a function of time.  This involves plotting and maths, and a quick search will point you to the package [matplotlib](https://matplotlib.org/index.html), and [this example](https://matplotlib.org/gallery/lines_bars_and_markers/simple_plot.html).  Here is that example:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(
    xlabel='time (s)',
    ylabel='voltage (mV)',
    title='About as simple as it gets, folks')
ax.grid()

plt.show()
```

### Help on specific functions
You might not know, or remember, what `np.arange` is.  Put the cursor somewhere on the word `arange` and press `shift-tab` to bring up a help window.  You can get more information by clicking on `+`.  Alternatively you could get this information in a code cell by typing `np.arange?` or `help(np.arange)`.

Keep your cells short.  This will aid debugging and improve the interactivity of your notebook.  You can split a long cell with `Ctrl Shift -` (or using the Edit menu in the task bar above) and then re-merge the parts once a problem is sorted out.

Now we'll edit and adapt the cell to make our plot:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
tau = 1
f = 10
s = np.exp(-t / tau) * np.sin(2 * np.pi * f * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(
    xlabel = 'time (s)',
    ylabel = r'$\exp(-t / \tau)\sin 2\pi f t$',
    title = 'An exponetially-decaying sine wave')
#ax.grid()
plt.show()
```

## My code won't stop!

Sometimes a cell never returns from execution. You see `In [*]:` to the left of it, and nothing is happening. It may be that the server is slow or you may have asked for a rather lengthy calculation, or you have asked for it to run without end.  Either way you want to stop it.  If you simply want to interrupt the kernel, click on the black square in the task bar.  Otherwise click on `Kernel` and choose an appropriate action. `Restart & Clear Output` will reset everything, and you can work through the cells again. `Restart & Run All` will also run the cells for you.  Every time you open a notebook it starts a new kernel so it's a good idea to `Shutdown` the kernel when you finish. This won't happen if you simply log out.
